# Matplotlib - Part 1

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

### Functional Method:

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()  # not required for jupyter notebook
```

#### Multiple plots on same canvas

```python
plt.subplot(1,2,1)  # no of rows, no of columns and subplot number
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Object Oriented method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])  #First arg: List [Left, bottom, width, height]

axes.plot(x,y)

axes.set_xlabel('X label')

axes.set_ylabel('Y label')

axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Outer Plot')

axes2.plot(y,x)
axes2.set_title('Inner Plot')
```
