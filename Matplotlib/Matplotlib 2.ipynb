## Figure Size and DPI

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
fig = plt.figure(figsize=(8,5))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,label='X square')

ax.plot(x, x**3, label='X cube')

ax.legend()

plt.show()
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,5),dpi=100)

axes[0].plot(x,y)

axes[1].plot(y,x) 

plt.tight_layout()
plt.show()
```

### Saving Figure
