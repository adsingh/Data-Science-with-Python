# Plot Appearence

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# color can be hex, linewidth == lw, linestyle == ls
ax.plot(x,y, color='#a21dc4', lw=4, alpha=0.8, ls='-', marker='o', markersize=20,
       markerfacecolor='green', markeredgewidth=5) 

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# color can be hex, linewidth == lw, linestyle == ls
ax.plot(x,y, color='#a21dc4', lw=4, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,3])


plt.show()
```
